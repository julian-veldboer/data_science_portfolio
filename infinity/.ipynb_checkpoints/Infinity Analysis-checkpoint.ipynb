```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('infinity.csv', delimiter=",", decimal=",")
data.drop([data.columns[9]], axis=1, inplace=True)
data.head()
```

```python
data.gain_per_minute.describe()
```

```python
plt.plot(data.Start, data.gain_per_minute)
plt.plot(data.Start, data.gain)
plt.title("gain analysis")
plt.xlabel("level")
plt.ylabel("gain")
plt.legend(["gain/min", "gain"])
plt.show()
```

```python
plt.plot(data.Start, data['time(s)'],'o')

# calc the trendline - linear
z = np.polyfit(data.Start, data['time(s)'], 1)
p = np.poly1d(z)
plt.plot(data.Start,p(data.Start),"r--")
# the line equation:
print ("linear \n y = %.6fx + %.6f" % (z[0], z[1]))

# calc the trendline - quadratic
z = np.polyfit(data.Start, data['time(s)'], 2)
p = np.poly1d(z)
plt.plot(data.Start,p(data.Start),"g--")
# the line equation:
print ("\nquadratic \n y = %.6fx2 + %.6fx1 + %.6f" % (z[0], z[1], z[2]))
plt.legend(["time","trend_linear", "trend_quad"])
plt.title("time anaylsis")
plt.xlabel("level")
plt.ylabel("time in s")
plt.show()
```
